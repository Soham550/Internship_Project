```python
'''1.	Data Collection & Exploration
	•	Load the Iris dataset (from sklearn.datasets or Kaggle).
	•	Perform exploratory data analysis (EDA) using Pandas, Matplotlib, and Seaborn.
	•	Visualize feature distributions and relationships using scatter plots and pair plots.'''
```

```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
file_path = "IRIS DATASET.xlsx"  # Change this to the correct file path if needed
df = pd.read_excel(file_path, sheet_name="Sheet4")

# Visualizing feature distributions and relationships
sns.set(style="whitegrid")
sns.pairplot(df, hue="FLOWER", diag_kind="kde", markers=["o", "s", "D"])
plt.show()
```

```python
'''2.Preprocessing & Feature Engineering
	•	Check for missing values and handle them if necessary.
	•	Normalize or standardize features (if required).
	•	Split the dataset into train and test sets (80-20 or 70-30).'''
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = df.drop(columns=["FLOWER"])
y = df["FLOWER"]

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split into train (80%) and test (20%) sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)

print("Feature standardization complete. Dataset split into training and testing sets.")
```

```python
'''3.	Model Training & Evaluation
	•	Train multiple classification models:
	•	Logistic Regression
	•	Decision Tree
	•	Random Forest
	•	Support Vector Machine (SVM)
	•	K-Nearest Neighbors (KNN)
	•	Compare their performance using accuracy, precision, recall, and F1-score.
	•	Use confusion matrices and ROC curves for evaluation.'''
	
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# models
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(probability=True),
    "KNN": KNeighborsClassifier()
}
for i,j in models.items():
    j.fit(X_train,y_train)
    y_pred=j.predict(X_test)
    print(f" {i} Model Evaluation:")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Classification Report:\n", classification_report(y_test, y_pred))

    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=y.unique(), yticklabels=y.unique())
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title(f"Confusion Matrix - {i}")
    plt.show()
```

```python
'''SVM performed the best (96.7% accuracy)'''
```

```python
'''4.	Model Optimization & Hyperparameter Tuning
	•	Perform GridSearchCV or RandomizedSearchCV to optimize hyperparameters.
	•	Implement cross-validation to improve model robustness.'''
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

# parameter grids for all models
param_grids = {
    "Logistic Regression": {
        "C": [0.01, 0.1, 1, 10, 100]
    },
    "Decision Tree": {
        "max_depth": [None, 5, 10, 20],
        "min_samples_split": [2, 5, 10]
    },
    "Random Forest": {
        "n_estimators": [10, 50, 100, 200],
        "max_depth": [None, 5, 10, 20],
        "min_samples_split": [2, 5, 10]
    },
    "SVM": {
        "C": [0.1, 1, 10, 100],
        "kernel": ["linear", "rbf", "poly"],
        "gamma": ["scale", "auto"]
    },
    "KNN": {
        "n_neighbors": [3, 5, 7, 9],
        "metric": ["euclidean", "manhattan", "minkowski"]
    }
}

# Models
models = {
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(probability=True),
    "KNN": KNeighborsClassifier()
}

# Train models using GridSearchCV
best_models = {}
for model_name, model in models.items():
    print(f"\nTuning {model_name}...")
    
    grid_search = GridSearchCV(model, param_grids[model_name], cv=5, scoring="accuracy", n_jobs=-1)
    grid_search.fit(X_train, y_train)

    best_models[model_name] = grid_search.best_estimator_
    print(f"Best {model_name} Parameters: {grid_search.best_params_}")
    print(f"Best {model_name} Cross-Validation Accuracy: {grid_search.best_score_:.4f}")

# Evaluate the best models on the test set
from sklearn.metrics import accuracy_score, classification_report

for model_name, best_model in best_models.items():
    y_pred = best_model.predict(X_test)
    print(f"\nFinal Evaluation - {model_name}:")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score

#models
models = {
    "Logistic Regression": LogisticRegression(C=10),
    "Decision Tree": DecisionTreeClassifier(max_depth=None, min_samples_split=2),
    "Random Forest": RandomForestClassifier(n_estimators=10, max_depth=5, min_samples_split=2),
    "SVM": SVC(C=0.1, kernel='linear', gamma='scale'),
    "KNN": KNeighborsClassifier(n_neighbors=9, metric='euclidean')
}

#cross-validation
cv_results = {}
for name, model in models.items():
    scores = cross_val_score(model, X_train, y_train, cv=5)  # 5-fold cross-validation
    cv_results[name] = scores.mean()
    print(f"{name} Cross-Validation Accuracy: {scores.mean():.4f}")
```
